#Presentación de funciones

Se tiene la serie $f(t) = 2\sum_{n=0}^{∞} \frac{(-1)^{n-1}}{n} \sin{t}$




Queremos calcular los primeros 50 armónicos de la serie, por lo que $f(t) = 2\sum_{n=0}^{50} \frac{(-1)^{n-1}}{n} \sin{t}$ \\

Nos damos cuenta que la serie está definida $\forall n \neq 0$

```python
def f(t, n):
    return 2*(-1)**(n-1)*np.sin(n*t)/n
```

```python
L = np.pi
n = 1
n_t = 50

t = np.linspace(-L, L, 1000)

F = 0
fig = plt.figure(figsize=(10,5))
plt.title(r'Serie de Fourier de f(t) = t, primeros {:.0f} armónicos'.format(n_t))
camera = Camera(fig)

while n < n_t:
    F+=f(t,n)
    plt.plot(t, F, c ='r')
    camera.snap()
    n+=1
```

```python
animation = camera.animate()
```
